### (2)  Kmeans 가 실패 : A Priori Algorithm 이용
- 장바구니 분석에서 주로 이용됨
- 아이템 집합 간의 연관성 강도를 측정하는 데 쓰는 신뢰도는 아래와 같이 조건절(AA)이 주어졌을 때 결과절(BB)이 일어날 조건부확률로 정의
- confidence(A→B)=P(A,B)/P(A)
- 향상도 : 생성된 규칙이 실제 효용가치가 있는지를 판별하는 데 사용   
    + 조건절과 결과절이 서로 독립일 때와 비교해 두 사건이 동시에 얼마나 발생하는지 비율   
    + 향상도가 1이라면 조건절과 결과절은 서로 독립 => 규칙 사이에 유의미한 연관성이 없음
    + 향상도가 2라면 두 사건이 독립이라는 걸 가정했을 때 대비 2배로 긍정적인 연관관계를 나타냅니다.
    + lift(A→B)=P(A,B)/(P(A)⋅P(B))   
https://ratsgo.github.io/machine%20learning/2017/04/08/apriori/

```python
#Apriori 를 위한 boolean sparse matrix 제작
import pandas as pd
# !pip install mlxtend --upgrade --no-deps
purchase=purchase==1 
purchase
# from mlxtend.frequent_patterns import association_rules

# print( association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3) ) 
```

```python
def cluster_apriori(purchase, threshold_lift):
    import mlxtend
    from mlxtend.frequent_patterns import apriori

    frequent_itemsets = apriori(purchase, min_support=0.00001, use_colnames=True)
    print(frequent_itemsets )
    # https://m.blog.naver.com/zzz90zzz/221807210555
    from mlxtend.frequent_patterns import association_rules

    df=pd.DataFrame(association_rules(frequent_itemsets, min_threshold=0.001))
    basket=df[df['lift']>threshold_lift]
    return basket

cluster_apriori(purchase, 2)
```

```python
def set_cluster(basket):
    item_sets=set()
    item_list=[]

    for _ in range(len(basket)):
        x=basket.iloc[_,0].union(basket.iloc[_,1])
        item_sets.add(x)
    for ele in item_sets:
        item_list.append(list(ele))
    return item_list
set_cluster(basket)
```

```python
from collections import defaultdict

def connected_components(lists):
    neighbors = defaultdict(set)
    seen = set()
    for each in lists:
        for item in each:
            neighbors[item].update(each)
    def component(node, neighbors=neighbors, seen=seen, see=seen.add):
        nodes = set([node])
        next_node = nodes.pop
        while nodes:
            node = next_node()
            see(node)
            nodes |= neighbors[node] - seen
            yield node
    for node in neighbors:
        if node not in seen:
            yield sorted(component(node))

L = set_cluster(basket)

print (list(connected_components(L)))
# https://stackoverflow.com/questions/4842613/merge-lists-that-share-common-elements
item_cluster=pd.DataFrame(list(connected_components(L)))
item_cluster=item_cluster.T
item_cluster.head(10)
```

```python
item_dict={}
for _ in range(len(list(connected_components(L)))):
    item_dict[_]=list(connected_components(L))[_]
item_dict #이거 추가하는 부분은 좀 나중에 해야겠다..
```
